What is an AI agent? How does it differ from a simple LLM API call?

An AI agent is a system that uses an LLM to iteratively decide actions, often calling external tools and maintaining state, to achieve a goal. Unlike a simple LLM API call, which is single-step and stateless, agents involve planning, tool use, and multi-step reasoning loops


An AI agent is a system that:

takes a goal (not just a prompt)
decides what steps to take
can use tools (APIs, search, DBs)
may loop until it reaches a result
Goal → Plan → Act → Observe → Repeat → Final Answer

Simple LLM API call
response = llm("Summarize this text")

👉 Characteristics:

single step
no memory (unless you pass it)
no external actions
deterministic flow


| Feature        | LLM Call | AI Agent   |
| -------------- | -------- | ---------- |
| Steps          | Single   | Multi-step |
| Control flow   | Fixed    | Dynamic    |
| Tool usage     | No       | Yes        |
| Memory         | Manual   | Built-in   |
| Reasoning loop | No       | Yes        |
| Goal-oriented  | Weak     | Strong     |

What is a tool-use pattern in agentic AI? How do you define tools for an LLM?

A “tool-use pattern” is how you let an LLM decide to call external functions (APIs, DB queries, calculators) instead of trying to answer everything from its own text generation. It’s the core capability that turns a model into something agent-like

Types of tools
1. API tools
weather, stock, search
2. Computation tools
calculator, code execution
3. Retrieval tools
vector DB, document search
4. Action tools
send email, book ticket


Common patterns
