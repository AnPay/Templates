```python
def matmul(m1,m2):
  result = [[0 for _ in range(len(m2[0]))] for _ in range(len(m1))]
  for i in range(len(m1)): #2
    for j in range(len(m2[0])):
      for k in range(len(m2)):
        result[i][j]+=m1[i][k]*m2[k][j]

  return result
```

What is operator overloading? How do you implement __add__ and __eq__?

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

    def __eq__(self, other):
        return self.x == other.x and self.y == other.y


p1 = Point(1, 2)
p2 = Point(3, 4)

p3 = p1 + p2     # uses __add__
p1 == p2         # uses __eq__
```

What are properties in Python? How do @property, @setter, and @deleter work?

Properties allow controlled access to attributes.

```python
class Person:
    def __init__(self, age):
        self._age = age

    @property
    def age(self):
        return self._age

    @age.setter
    def age(self, value):
        if value < 0:
            raise ValueError("Invalid age")
        self._age = value

    @age.deleter
    def age(self):
        del self._age
```

Linked list

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insert(self, val):
        new = Node(val)
        new.next = self.head
        self.head = new

    def print_list(self):
        curr = self.head
        while curr:
            print(curr.val)
            curr = curr.next
```

BST

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = self.right = None

def insert(root, val):
    if not root:
        return Node(val)
    if val < root.val:
        root.left = insert(root.left, val)
    else:
        root.right = insert(root.right, val)
    return root

def search(root, val):
    if not root or root.val == val:
        return root
    if val < root.val:
        return search(root.left, val)
    return search(root.right, val)
```

Singleton

```python
class Singleton:
    _instance = None

    def __new__(cls):
        if not cls._instance:
            cls._instance = super().__new__(cls)
        return cls._instance
```

What is a vector database?

A vector database stores high-dimensional vectors (embeddings) and lets you query by similarity.

Typical flow:
Convert data → embedding (e.g., sentence → vector)
Store (id, vector, metadata)
Query with a new vector
DB returns nearest neighbors (most similar items)
Why it matters
Semantic search (meaning, not keywords)
RAG (retrieve context for LLMs)
Recommendations / dedup / clustering

Under the hood: ANN (Approximate Nearest Neighbor) algorithms like HNSW/IVF for fast search.

How do you implement a semantic search system using sentence-transformers?

“I use sentence-transformers to encode documents and queries into embeddings, then compute cosine similarity to retrieve top-k results. For scalability, I use FAISS or a vector database instead of brute-force search. I also normalize embeddings and chunk documents to improve retrieval quality.”

```python
def semantic_search(query, docs, model):
    doc_emb = model.encode(docs, normalize_embeddings=True)
    q_emb = model.encode(query, normalize_embeddings=True)

    scores = doc_emb @ q_emb
    idx = scores.argsort()[::-1]

    return [docs[i] for i in idx[:3]]
```
